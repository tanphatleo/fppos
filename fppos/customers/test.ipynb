```python
import pandas as pd
```

```python
data_province = pd.read_excel('addresses.xlsx', sheet_name='province')
```

```python
json_ = []
for index, row in data_province.iterrows():
    item = {}
    item['model'] = 'customers.Province'
    item['pk'] = int(row['id'])
    fields = {
        'name': row['name'],
        'is_active': True
    }
    item['fields'] = fields
    json_.append(item)
```

```python
# write 
with open('provinces.json', 'w', encoding='utf-8') as f:
    import json
    json.dump(json_, f, ensure_ascii=False, indent=4)
```

```python
data_wards = pd.read_excel('addresses.xlsx', sheet_name='ward')
```

```python
json_wards = []
for index, row in data_wards.iterrows():
    item = {}
    item['model'] = 'customers.Ward'
    item['pk'] = int(row['id'])
    fields = {
        'name': row['ward'],
        'province_id': int(row['provincei_id']),
        'is_active': True
    }
    item['fields'] = fields
    json_wards.append(item)
# write
with open('wards.json', 'w', encoding='utf-8') as f:
    import json
    json.dump(json_wards, f, ensure_ascii=False, indent=4)
```

```python
data_cust = pd.read_excel('customers.xlsx')
```

```python
json_cust = []
for index, row in data_cust.iterrows():
    item = {}
    item['model'] = 'customers.Customer'
    item['pk'] = int(row['id'])
    fields = {
        'name': row['name'],
        'code': row['code'],
        'phone_number': str(row['phone_number']),
        'address': row['address'],
        'province_id': None if pd.isna(row['province_id']) else int(row['province_id']),
        'ward_id': None if pd.isna(row['ward_id']) else int(row['ward_id']),
        'is_active': True,
        'created_at': "2024-01-01T00:00:00Z",
        'updated_at': "2024-01-01T00:00:00Z"
    }
    item['fields'] = fields
    json_cust.append(item)
```

```python
# write
with open('customers.json', 'w', encoding='utf-8') as f:
    import json
    json.dump(json_cust, f, ensure_ascii=False, indent=4)
```
