```python
import pandas as pd
```

```python
data = pd.read_excel('product.xlsx')
print(data.head())
```

```python
json_ = []
id = 1
for index, row in data.iterrows():
    temp_ = {
        "model": "products.product",
        "pk": id,
        "fields": {}
    }
    for col in data.columns:
        # if row[col] is nan then set it to null
        if pd.isna(row[col]):
            temp_["fields"][col] = None
        elif col == "package_details":
            output = row[col].split('|')
            array = []
            for i in range(len(output)):
                # before : split
                item = output[i].split(':')
                array.append({
                    "code": item[0],
                    "quantity": int(item[1])
                })

            temp_["fields"][col] = array
        else:
            temp_["fields"][col] = row[col]
        temp_["fields"]["created_at"] = "2024-07-01T00:00:00Z"
        temp_["fields"]["updated_at"] = "2024-07-01T00:00:00Z"
    json_.append(temp_)
    id += 1
```

```python
# save to json file
with open('products_fixture.json', 'w') as f:
    import json
    json.dump(json_, f, indent=4)
```

```python
# read and find unquie product_group values
unique_product_groups = data['product_group'].unique()
print(unique_product_groups)
```

```python
# make json fixture for product groups
for i, group in enumerate(unique_product_groups):
    group_json = {
        "model": "products.productgroup",
        "pk": i + 1,
        "fields": {
            "name": group
        }
    }
    print(group_json)
```

```python
# save to json file
with open('product_groups_fixture.json', 'w') as f:
    group_fixtures = []
    for i, group in enumerate(unique_product_groups):
        group_json = {
            "model": "products.productgroup",
            "pk": i + 1,
            "fields": {
                "name": group,
                "created_at": "2024-07-01T00:00:00Z",
                "updated_at": "2024-07-01T00:00:00Z"
            }
        }
        group_fixtures.append(group_json)
    json.dump(group_fixtures, f, indent=4)
    # json.dump(group_fixtures, f, indent=4)
```
